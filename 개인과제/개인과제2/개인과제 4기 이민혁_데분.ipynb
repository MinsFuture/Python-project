```python
## 2-2 상위 3개 제품군의 월렌탈비용 합계 계산 : 제품군(Product_Type) 중 계약 건수(빈도)가 가장 높은 상위 3개의 항목을 추출하여 df2로 선언한 뒤, 
## 계약일자(Datetime)에 따른 월렌탈비용(Amount_Month)의 합계를 계산하고 result5.xlsx 파일로 저장하시오.

top_product_types = df['Product_Type'].value_counts().nlargest(3).index

# 상위 3개 제품군에 해당하는 데이터 필터링
df2 = df[df['Product_Type'].isin(top_product_types)]
print(df2.groupby('Datetime')['Amount_Month'].sum())
df2.groupby('Datetime')['Amount_Month'].sum().to_excel('result5.xlsx')
```

```python
## 4-1 계약 연도 파생 변수 생성 및 월렌탈 비용 합계 : 계산계약일자(Datetime)에서 계약 연도(Year) 파생 변수를 생성하고, 
## 연도별로 각 제품군(Product_Type)의 월렌탈비용(Amount_Month) 합계를 계산하시오.

df['Year'] = pd.to_datetime(df['Datetime']).dt.year

df.groupby(['Year', 'Product_Type'])['Amount_Month'].sum()
```

```python
## 4-2 제품군별 증감율 계산 (2022년 대비 2023년) : 2022년과 2023년 각 제품군(Product_Type)별 월렌탈비용(Amount_Month) 합계의 증감율을 계산하시오.
df4_2 = df.groupby(['Year', 'Product_Type'])['Amount_Month'].sum().to_frame()
df4_2 = (df4_2.loc[2022] - df4_2.loc[2023]) / df4_2.loc[2022]
df4_2.columns = ['rate']

df4_2
```

```python
## 4-3 판매 유형별 월렌탈비용 평균 차이 계산 : 판매 유형(Sales_Type)에서 '렌탈' 계약과 '멤버십' 계약의 월렌탈비용(Amount_Month) 평균 차이를 계산하시오.
df4_3 = df.groupby('Sales_Type')['Amount_Month'].mean()

print(' 차이 : ', abs(df4_3['렌탈'] - df4_3['멤버십']))
```

```python
## 4-4 주차별 계약 데이터 자동 정리 함수 생성계약 데이터를 주기적으로 정리하는 업무 자동화 프로그램을 위해, 
## 데이터에서 주차(Work Week)를 입력받아 해당 주차의 제품군(Product_Type)별 월렌탈비용(Amount_Month)의 합계와 평균을 엑셀 파일로 출력하는 함수를 생성하시오.

df['Datetime'] = pd.to_datetime(df['Datetime'])
df['WorkWeek'] = df['Datetime'].dt.isocalendar().week

def auto_week(week):
    df4_4 = df[df['WorkWeek'] == week]
    result = pd.DataFrame({
        '합계' : df4_4.groupby('Product_Type')['Amount_Month'].sum(),
        '평균' : df4_4.groupby('Product_Type')['Amount_Month'].mean()
    })

    result.to_excel('result4_4.xlsx')
    print(result)

input_week = int(input('주차를 입력하세요 : '))

auto_week(input_week)
```
