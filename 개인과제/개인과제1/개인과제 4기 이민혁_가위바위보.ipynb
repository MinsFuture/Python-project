```python
# [개인과제] 4기 이민혁_가위바위보
import pandas as pd
import pickle
import random 

def init_data():
    try:
        with open("user.dat", 'rb') as f: 
            return pickle.load(f)
    except (FileNotFoundError, EOFError):
        data = []
        with open("user.dat", "wb") as f:
            pickle.dump(data, f)
        return data

users = init_data()
print('가위바위보 게임에 오신걸 환영합니다')
flag = 0

def rps_game_start():
    global users
    name = input('사용자의 이름을 입력하세요 : ')
    score = 0
    win_straight_score = 1 # 몇 연승인지 체크하는 변수

    while True:  
        user_rps = input('가위, 바위, 보 중 하나를 입력 해 주세요 : ')
        if not user_rps in ['가위', '바위', '보']:
            print(' ========== [가위, 바위, 보] 중 하나를 정확히 입력 해 주세요  : ========== ')
            continue
        com_cps = random.choice(['가위', '바위', '보'])
    
        # 가위바위보를 이기거나 지면
        if user_rps != com_cps:
            # 이기면 2배
            if ((user_rps == '가위') & (com_cps == '보')) | ((user_rps == '바위') & (com_cps == '가위')) | ((user_rps == '보') & (com_cps == '바위')):
                score += win_straight_score
                win_straight_score *= 2
                print('이겼습니다!, 현재 점수는 : ', score)
        
            # 지면 그대로 랭킹 수정하고 저장, 출력 후 종료
            else:
                print('졌습니다 ...')
                print('========== 랭킹 시작 ==========')
                data = {'이름' : name, '점수' : score, '랭킹' : 0}
                users.append(data)
                users.sort(key=lambda x: (-x['점수'], x['이름']))
                rank = 1
                for element in users:
                    element['랭킹'] = rank
                    print(f'{element['랭킹']}등 : {element['이름']}, 점수 : {element['점수']}')
                    rank += 1
                print('========== 랭킹 끝 ==========')
                f = open("user.dat", "wb")
                pickle.dump(users, f)        
                f.close()
                break
        else:
            print('비겼습니다, 현재 점수는 : ', score)
            win_straight_score = 1
    
def get_rank():
    global users
    print('========== 랭킹 시작 ==========')
    for element in users:
        print(f'{element['랭킹']}등 : {element['이름']}, 점수 : {element['점수']}')
    print('========== 랭킹 끝 ==========')
    
while True:   
    print(""" 
        1. 게임 시작
        2. 랭킹 조회
        3. 게임 종료
    """)

    flag = int(input('번호를 입력하세요 : '))
    if flag == 1:
        rps_game_start()
    elif flag == 2:
        get_rank()   
    elif flag == 3:
        print('게임 종료')
        break;
    
```
