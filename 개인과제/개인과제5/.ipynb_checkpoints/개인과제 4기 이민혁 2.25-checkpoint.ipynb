4. **연령대별 분석 및 평균 차이 계산**
    - **목표:**
        - '연령' 데이터를 바탕으로 고객의 '연령대'를 산출(예: 20대, 30대 등)
        - 각 연령대의 '총구매금액' 합계를 계산한 후, 합계가 가장 높은 두 개의 연령대를 선택
        - 두 연령대 간의 '총구매금액' 평균의 차이를 산출

```python
df2['연령'].min(), df2['연령'].max() 
df2['연령대'] = (df2['연령'] // 10 * 10).astype(str) + '대'

first, second = df2.groupby('연령대')['총구매금액'].sum().sort_values(ascending=False).head(2).index

first_mean = df2[df2['연령대'] == first]['총구매금액'].mean()
second_mean = df2[df2['연령대'] == second]['총구매금액'].mean()

first_mean - second_mean
```

5. **회원상태별 총구매금액 평균 시각화**
    - **목표:** '회원상태'에 따른 '총구매금액'의 평균을 시각화 (예: Bar Chart 등)
    - **참고:** Matplotlib, Seaborn 등 시각화 라이브러리 활용

```python
df = df2.groupby('회원상태')['총구매금액'].mean().to_frame()

plt.bar(df.index, df['총구매금액'], alpha = 0.5, edgecolor = 'black')
plt.xlabel('회원상태')
plt.ylabel('총구매금액(백만원)')
```

6. **1회 방문 구매 평균 및 고객 등급 분류**
    - **계산:** '총구매금액'을 '총방문횟수'로 나누어 '1회방문구매평균'을 산출
    - **분류 기준:**
        - 1회 방문 구매 평균 ≥ 10만원 → **A 등급**
        - 1회 방문 구매 평균 ≥ 5만원 → **B 등급**
        - 1회 방문 구매 평균 ≥ 2만원 → **C 등급**
        - 그 외 → **D 등급**
    - **출력:** 분류 결과를 새로운 파생변수 '고객등급'으로 추가

```python
df2['1회방문구매평균'] = df2['총구매금액'] / df2['총방문횟수']

def get_classify(mean):
    if mean >= 100000:
        return 'A'
    elif mean >= 50000:
        return 'B'
    elif mean >= 20000:
        return 'C'
    else:
        return 'D'

df2['고객등급'] = df2['1회방문구매평균'].apply(get_classify)
df2
```

7. **기혼자가 가장 많은 고객 등급 확인**
    - **목표:** 새로 생성된 '고객등급' 변수에서, 기혼 고객의 수가 가장 많은 등급을 파악

```python
df_married = df2[df2["결혼유무"] == "기혼"]
df = df_married.groupby('고객등급')['결혼유무'].value_counts().to_frame()
df['count'].idxmax()[0]
```

8. **지역 파생변수 생성 (주소 기반 시 분류)**
    - **목표:** `df2`의 '주소' 데이터를 활용하여 "지역(시)"라는 새로운 파생변수를 생성
    - **분류 기준:**
        - **서울시, 용인시, 성남시, 수원시, 광주시:** 해당 시명을 그대로 사용
        - **나머지 지역:** "기타"로 분류

```python
def get_region(region):
    region = region.split(" ")
    result = ''
    for data in region:
        if data == '서울':
            result = '서울시'
            break
        elif data == '용인시':
            result = '용인시'
            break
        elif data == '성남시':
            result = '성남시'
            break
        elif data == '수원시':
            result = '수원시'
            break
        elif data == '광주시':
            result = '광주시'
            break
        else:
            result = '기타'
    return result

df2['지역(시)'] = df2['주소'].apply(get_region)
df2
```

9. **고객 등급별 성별 총구매금액 기술통계량 저장 프로그램**
    - **목표:**
        - 사용자로부터 특정 '고객등급'을 입력받아 해당 등급 내 '성별'에 따른 '총구매금액'의 기술통계량(평균, 표준편차 등)을 계산
        - 계산 결과를 `result9_등급명.xlsx` (여기서 "등급명"은 입력한 고객 등급) 파일로 저장
