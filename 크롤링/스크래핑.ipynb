```python
# 데이터 스크래핑 : 인터넷에서 정보를 가져오는 기술
import requests # 페이지 요청
from bs4 import BeautifulSoup # html 소스코드를 파이썬으로 다룰 수 있게 변환
import time 
import pandas as pd

url = requests.get('https://search.daum.net/search?w=tot&DA=YZR&t__nil_searchbox=btn&q=%EB%A1%9C%EB%98%90')
```

```python
def make_numbers(html):
    box = html.find('div', class_ = 'lottonum').find_all('span')
    del box[6:8]
    box

    numbers = []
    for i in box:
        numbers.append(int(i.text))
    return numbers
    
```

```python
make_numbers(html)
```

```python
total = []
for i in range(1, 21):
    url = requests.get(f'https://search.daum.net/search?w=tot&rtmaxcoll=LOT&DA=LOT&q={i}%ED%9A%8C%20%EB%A1%9C%EB%98%90%EB%8B%B9%EC%B2%A8%EB%B2%88%ED%98%B8')
    html = BeautifulSoup(url.text)
    numbers = make_numbers(html)
    total.append(numbers)
    print(f'{i}회 로또 데이터 저장완료 : {numbers}')
    time.sleep(1)
```

```python
lotto = pd.DataFrame(total, columns=['1번째', '2번째', '3번째', '4번째', '5번째', '6번째', '보너스'])
lotto.index = [f"{i}회차" for i in range(1, 21)]
lotto.to_excel('lotto.xlsx')
lotto
```

```python
def get_numbers(html):
        temp = html.find('tr', class_ = 'fst')
        span_tags = temp.find('td', class_ = 'number').find_all('span')
    
        # 몇조인지
        team = temp.find('td', class_ = 'each').text
        total = [team]
        for i in range(0, 6):
            total.append(int(span_tags[i].text))
        
        return total
    
result = []

for i in range(1, 21):
    time.sleep(2)
    url = requests.get(f'https://search.daum.net/search?w=tot&DA=JIM&rtmaxcoll=JIM&&q=%EC%97%B0%EA%B8%88%EB%B3%B5%EA%B6%8C%20720%2B%20{i}%ED%9A%8C%EC%B0%A8')
    html = BeautifulSoup(url.text)
    numbers = get_numbers(html)
    print(numbers)
    result.append(numbers)
```

```python
df = pd.DataFrame(result, columns = ['조'] + [f'{i}번째' for i in range(1, 7)])
df.to_excel('연금복권.xlsx')

df
```
