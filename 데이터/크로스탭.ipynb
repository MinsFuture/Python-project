```python
# CrossTab 빈도값
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
pd.crosstab(index = tips['sex'], columns=tips['day'], 
           values = tips['tip'], aggfunc = 'sum', margins=True, margins_name='총계', normalize='columns')
```

```python
# pivot table : 내가 원하는 행과, 열을 설정해서 집계값
import warnings
warnings.filterwarnings('ignore')

tips.pivot_table(index=['sex', 'time'], columns='day', values='tip', aggfunc=['sum', 'mean'], fill_value=0, margins=True)
```
